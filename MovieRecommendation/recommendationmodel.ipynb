```python
# importing dependecies

import numpy as np 
import pandas as pd 
import difflib 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#data collection and pre-processing

df = pd.read_csv('movies.csv')
df.head()
```

```python
df.shape
```

```python
# feature selection

selected_features = ['genres','keywords','tagline','cast','director']
selected_features
```

```python
# replacing null values

for feature in selected_features:
    df[feature] = df[feature].fillna('')
```

```python
#combining selected features

combined_features = df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
combined_features
```

```python
# converting text data to feature vectors

vectorizer = TfidfVectorizer()

feature_vectors = vectorizer.fit_transform(combined_features)
feature_vectors
```

```python
# similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)

similarity
```

```python
# getting movie name from user

movie_name = input('Enter your favourite movie name : ')
```

```python
#creating a list with all the movie names

list_of_all_titles = df['title'].tolist()
```

```python
# finding match for name given by user

find_match = difflib.get_close_matches(movie_name, list_of_all_titles)
find_match
```

```python
close_match = find_match[0]
close_match
```

```python
# finding the index of movie with title

movie_index = df[df.title == close_match]['index'].values[0]
movie_index
```

```python
#getting list of similar movies

similarity_score = list(enumerate(similarity[movie_index]))
similarity_score
```

```python
#sorting movies based on similarity score

sort_movies = sorted(similarity_score, key = lambda x:x[1],reverse=True)
sort_movies
```

```python
# give recommendations for movies
print('Movies suggested for you: \n')

i = 1

for movie in sort_movies:
    index = movie[0]
    title_from_index = df[df.index == index]['title'].values[0]
    if(i<11):
     print(i,'.',title_from_index)
     i+=1
```
