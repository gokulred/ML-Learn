```python
X=df[['number of bedrooms', 'number of bathrooms','living area','condition of the house','Number of schools nearby']]
Y=df[['Price']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from  sklearn.tree import DecisionTreeRegressor
param_grid = {
    "criterion" : ["mse","friedman_mse","mae"],
    "splitter"  : ["best","random"],
    "max_depth" : [None ,10, 20, 30, 40, 50],
    "min_samples_split" : [2,5,10],
    "min_samples_leaf"  :[1,2,4]
}
```

```python
tree_model = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator = tree_model, param_grid = param_grid)
```

```python
grid_tree.fit(X_train,Y_train)
```

```python
grid_tree.best_params_
```

```python
tree_preds = grid_tree.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(Y_test,tree_preds)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
predslr=lr.predict(X_test)
```

```python
mean_absolute_error(Y_test,predslr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
param_gridrf = {
    "max_depth" : [5,10,15],
    "n_estimators" : [2,3,4,5,6,7,8,]
}
```

```python
gridrf = GridSearchCV(rf, param_gridrf)
```

```python
gridrf.fit(X_train , Y_train.values.ravel())
```

```python
gridrf.best_params_
```

```python
rfprediction = gridrf.predict(X_test)
```

```python
mean_absolute_error(Y_test,rfprediction)
```

```python
import joblib
joblib.dump(gridrf,"model.pkl")
```
